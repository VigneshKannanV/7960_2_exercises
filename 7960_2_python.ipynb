```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df1=pd.read_csv('imdb.csv',escapechar='\\')
df2=pd.read_csv('movie_metadata.csv')
```

```python
l=[]
for i in df1.itertuples():
    l.append(len(i[3])-7)
df1['TitleLength']=l 
```

```python
sns.set_style('whitegrid')
```

Imdb Datset

```python
sns.jointplot(x='imdbRating',y='duration',data=df1,kind='hex',color='tomato',ylim={0,20000}).set_axis_labels("Rating_of_the_movie", "Duration_of_the_movie")
```

```python
sns.rugplot(df1['duration'])
```

```python
sns.barplot(x='type',y='nrOfWins',data=df1,hue='Animation')
```

```python
plt.figure(figsize=(18,3))
sns.countplot(x='year',data=df1)
plt.tight_layout()
```

```python
sns.boxplot(x='type',y='TitleLength',data=df1)
```

```python
sns.stripplot(x='type',y='duration',data=df1,jitter=True,hue='Comedy')
```

```python
df1c=df1.corr()
```

```python
sns.heatmap(df1c,cmap='YlGnBu', linewidths=0.5 ,linecolor='white')
```

```python
g=sns.FacetGrid(data=df1,col='Action',row='Crime')
g.map(sns.distplot,'imdbRating')
```

```python
sns.lmplot(x='ratingCount',y='nrOfUserReviews',data=df1).set_axis_labels('Ratings_Count','Total_no_of_Users_Reviewed')
```

Movie_Metadata Dataset

```python
sns.jointplot(x='gross',y='imdb_score',data=df2,kind='reg',xlim={0,100000000},color='lawngreen')
```

```python
sns.kdeplot(df2['imdb_score'])
```

```python
plt.figure(figsize=(18,3))
sns.barplot(x='content_rating',y='gross',data=df2,palette='dark')
plt.tight_layout()
```

```python
plt.figure(figsize=(18,3))
sns.violinplot(x='content_rating',y='imdb_score',data=df2,palette='pastel')
plt.tight_layout()
```

```python
plt.figure(figsize=(18,3))
sns.swarmplot(x='aspect_ratio',y='duration',data=df2)
plt.tight_layout()
```

```python
sns.set_context('paper')
sns.lmplot(x='title_year',y='duration',data=df2,palette='coolwarm')
```
